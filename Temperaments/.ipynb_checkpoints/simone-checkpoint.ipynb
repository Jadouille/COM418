```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import IPython

import matplotlib
plt.rcParams["figure.figsize"] = (14,4)

import ipywidgets as widgets

def play(clip, rate=SF):
    display(IPython.display.Audio(clip, rate=rate))
    

SF = 32000  
```

```python
def click(N, SF):
    p = np.zeros(N)
    p[0] = 1
    return np.tile(p, int(SF / N))
```

```python
c = click(10000, 2*SF)

plt.plot(c);
plt.show()

plt.plot(np.abs(np.fft.fft(c)));
plt.show()

play(c, SF)
```

```python
c = click(1000, 2*SF)
sos = signal.butter(10, [50 / SF, 100 / SF], 'bandpass', output='sos')
cf = signal.sosfilt(sos, c)
```

```python
plt.plot(cf);
plt.show()

plt.plot(np.abs(np.fft.fft(cf)));
plt.show()

play(cf, SF)
```

```python
def infra(K, L, FS):
    N = FS * L
    x = np.zeros(N, dtype=float)
    for n in np.linspace(12, 24, K):
        x += np.cos(2 * np.pi * n / FS * np.arange(0, N) + np.random.uniform(0, 6.28))
    return x
        
```

```python
c = infra(200, 2, SF)
plt.plot(c)
play(c, SF)
```
